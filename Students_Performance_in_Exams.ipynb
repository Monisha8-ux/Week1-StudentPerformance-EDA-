```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
df.isnull().sum()
df.dropna(inplace=True)
```

```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
```

Gender vs Scores


```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
sns.barplot(x='gender', y='math score', data=df)
plt.title('Math Score by Gender')
```

Parental level of education


```python
df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.xticks(rotation=45)
plt.title('Math Score vs Parental Education')
```

Lunch Type Impact


```python
sns.boxplot(x='lunch', y='reading score', data=df)
plt.title('Reading Score by Lunch Type')
```

Test Preparation Course Effect


```python
sns.barplot(x='test preparation course', y='writing score', data=df)
plt.title('Writing Score by Test Preparation Course')
```

Distribution of Scores


```python
sns.histplot(df['math score'], kde=True)
plt.title('Distribution of Math Scores')
```

## Insights from Student Performance Analysis

1. Gender & Scores: Female students scored higher on average in reading (74 vs 69) and writing (76 vs 65) compared to males, while males slightly outperformed in math (69 vs 68).

2. Lunch Type Impact: Students with standard lunch achieved ~10 points higher in all subjects than those with free/reduced lunch.

3. Test Preparation Course: Those who completed the prep course scored ~812 points higher across all subjects.

4. Parental Education: Higher parental education (bachelors or masters) correlates with better student performance, especially in reading.

5. Score Distributions: Math scores showed greater variation, while reading and writing were more consistent across students.
